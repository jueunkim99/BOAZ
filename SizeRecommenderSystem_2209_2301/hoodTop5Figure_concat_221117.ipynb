```python
#(22FW) 2 TONE ARCH HOODIE GRAY: https://www.musinsa.com/app/goods/2326935
#toffee_에센셜 후드 스웻 셔츠 (MELANGE GREY)(4개): https://www.musinsa.com/app/goods/2314616
#signlogo_black: https://www.musinsa.com/app/goods/644025
#signlogo_gray(4개): https://www.musinsa.com/app/goods/644026
#lee(빅 트위치 루즈핏 후드 티셔츠 네이비):https://www.musinsa.com/app/goods/1571501
#(22FW) 2 TONE ARCH HOODIE GRAY/ 에센셜 후드 스웻 셔츠 (BLACK)/ SIGN LOGO HOOD BLACK/  SIGN LOGO HOOD GRAY/ 빅 트위치 루즈핏 후드 티셔츠 네이비

options = webdriver.ChromeOptions()
options.add_experimental_option("excludeSwitches", ["enable-logging"])
driver = webdriver.Chrome(options=options)
driver.get('https://www.musinsa.com/app/goods/2326935')
time.sleep(5)

html = driver.page_source
soup = BeautifulSoup(html,'html.parser')
figure = soup.find('table',{'class':'table_th_grey'})
p = parser.make2d(figure)
figure_df1 = pd.DataFrame(data = p[1:],columns = p[0])
figure_df1.drop([0,1],inplace = True)
```

```python
df1 = pd.read_csv('hood1.csv',encoding = 'cp949')
df2 = pd.read_csv('hood2.csv')#,encoding = 'cp949')
df3 = pd.read_csv('hood3.csv')#,encoding = 'cp949')
df4 = pd.read_csv('hood4.csv')#,encoding = 'cp949')
df5 = pd.read_csv('hood5.csv')#,encoding = 'cp949')
```

```python
merge_df1 = pd.merge(df1,figure_df1,how = 'left',left_on = 'size',right_on = 'cm')
merge_df2 = pd.merge(df2,figure_df2,how = 'left',left_on = 'size',right_on = 'cm')
merge_df3 = pd.merge(df3,figure_df3,how = 'left',left_on = 'size',right_on = 'cm')
merge_df4 = pd.merge(df4,figure_df4,how = 'left',left_on = 'size',right_on = 'cm')
merge_df5 = pd.merge(df5,figure_df5,how = 'left',left_on = 'size',right_on = 'cm')
```
